```python
#Exercise 11, Outlier detection
import pandas as pd
df = pd.DataFrame({
"salary":[3500,4000,4200,3900,4100,45000]
})
print(df)
Q1 = df["salary"].quantile(0.25)

Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]

print(outliers)

clean_df = df[
(df["salary"] >= lower) &
(df["salary"] <= upper)
]

print(clean_df)
```

```python
#Exercise 12, Phone number standardization
import pandas as pd
df = pd.DataFrame({
"phone":[
"0911223344",
"+251911223344",
"251911223344",
"0911-223-344"
]
})
print(df)
df["phone"] = df["phone"].str.replace("-", "", regex=False)

df["phone"] = df["phone"].str.replace("+", "", regex=False)
df["phone"] = df["phone"].apply(
lambda x:
"+251" + x[1:]
if x.startswith("0")
else "+251" + x[3:]
)

print(df)
```

```python
#Exercise 15, Mixed Data Cleaning
import pandas as pd
import numpy as np


df = pd.DataFrame({
"name":[" Abel ","Sara",None,"JOHN"],
"age":[20,np.nan,25,30],
"salary":[4000,4500,5000,45000],
"join_date":[
"2026-01-01",
"01/02/2026",
"March 10, 2026",
"2026.04.15"
]
})
print(df)
df["name"] = (
df["name"]
.fillna("Unknown")
.str.strip()
.str.title()
)

df["age"] = df["age"].fillna(df["age"].mean())

df["join_date"] = pd.to_datetime(df["join_date"])
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]

print(df)

print("Outliers")

print(outliers)
```
