```python
import pandas as pd
import numpy as np
university={"dtudent_id":[201,202,203, 204, 205,206,207, 208, 209,210],
            "name":["Abel","Sara",None,"John","Marta",None,"David","Helen","Tom",None],
            "department":["CS","CS", "CS",None,"SE", "IT","SE", None,"CS","IT"],
            "gpa":[3.5, 3.8, np.nan,2.9, 3.7, np.nan, 3.2, 3.4, 3.9, np.nan],
            "scholarship":[5000,np.nan, 3000, 2000, np.nan, 4000, 3500, np.nan, 5000, np.nan]
           }
#Creating the data frame using pandas
university_df=pd.DataFrame(university)
#Counting empty values
total_empty = university_df.isna().sum().sum()
print(f"Total empty values: {total_empty}")
#geting the total dataframe size
total_size=university_df.size
#Calculate the percentage of the empty values
percentage=(total_empty/total_size) *100
print(f"Missing values perecentage:{percentage}")
#Selecting columns with ob ject or string and numeric type
text_data=university_df.select_dtypes(include=['object']).copy()
numeric_data=university_df.select_dtypes(include=['number']).copy()
text_data=text_data.fillna("Unknown")
gpa_median=numeric_data["gpa"].median()
numeric_data["gpa"]=numeric_data["gpa"].fillna(gpa_median)
# replacing missing scholarship with  zero
numeric_data['scholarship']=numeric_data["scholarship"].fillna(0)
#Replacing the department with most frequent values
col_mode = text_data[text_data['department'] != "Unknown"]['department'].mode().iloc[0]
text_data['department'] = text_data['department'].replace("Unknown", col_mode)
#count studnt by department
valid_students = university_df[university_df['name'].notna()]
dept_counts = valid_students['department'].value_counts(dropna=False)
print(f"Totoal number of studnets availale based on the department:\n{dept_counts}")
print("*"*30)
has_scholarship=university_df[university_df['scholarship']>0]
dept_scholarship=has_scholarship.groupby('department').size().reset_index(name='count_student')
print(dept_scholarship)
print("*"*30)
print(f"total :{dept_scholarship['count_student'].sum()} studnets take scholarship")
top_student_row = university_df.loc[university_df['gpa'].idxmax()]
bottom_student_row = university_df.loc[university_df['gpa'].idxmin()]
print(f"Top Student is {top_student_row['name']} with a {top_student_row['gpa']}")
print(f"Studnet with low gpa is {bottom_student_row['name']} with a {bottom_student_row['gpa']}")

# calculationg average of gpa
average_gpa=university_df['gpa'].mean()
print(f"Average of gpa column :{round(average_gpa, 2)}")
#Calcualting total scholarship amount
scholarship_amount=university_df['scholarship'].sum()
print(f"Total amount of scholarship: {scholarship_amount}")
print("*"*30)
print("Final Data structure")
df_final = pd.concat([text_data, numeric_data], axis=1)
print(df_final)

# print(university_df.dtypes) 
# print(university_df)
# print(university)
    
```
