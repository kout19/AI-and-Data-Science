```python
#Section D: Data Cleaning  Exercise 19: Handling Missing Values
import pandas as pd
import numpy as np
data = {
    "Student Name": ["Alice", "Bob", "Charlie", "Diana", "Evan"],
    "Age": [20, np.nan, 19, 22, np.nan],
    "Score": [85.5, 92.0, np.nan, 88.0, 45.0],
    "Attendance": ["Present", "Present", "Absent", None, "Present"]
}
df = pd.DataFrame(data)
df_mean=df.copy()
df_mean['Age']=df_mean['Age'].fillna(df_mean['Age'].mean())
df_mean['Score']=df_mean['Score'].fillna(df_mean['Score'].mean())
df_median=df.copy()
df_median['Age']=df_median['Age'].fillna(df_median['Age'].median())
df_median['Score']=df_median['Score'].fillna(df_median['Score'].median())
print(f"Mean values for numerical data:\n{df_mean}")
print("*"*50)
print(f"Median values for numerical data :\n{df_median}")
print("*"*50)
print(df)
```

```python
#Section D: Data Cleaning  Exercise 20, Duplicate Records
import pandas as pd
data = {
    "Student Name": ["Alice", "Bob", "Charlie", "Alice", "Diana", "Bob"],
    "Age": [20, 21, 19, 20, 22, 21],
    "Score": [85.5, 92.0, 78.5, 85.5, 88.0, 92.0]
}

df = pd.DataFrame(data)
print("Original DataFrame with Duplicates:")
duplicate_mask = df.duplicated()
print("Duplicate Mask (True means the row is a duplicate):")
print(duplicate_mask)
print("\nIsolated Duplicate Rows:")
print(df[duplicate_mask])
print(df)
```

```python
#Section D: Data Cleaning Exercise 22: Correcting Data Entry Errors
import pandas as pd
import numpy as np
ages = pd.Series([21, 25, 230, 30, 19, -5, 27])
cleaned_ages = pd.Series(np.where((ages < 0) | (ages > 120), np.nan, ages))

print("Original Ages:")
print(ages.to_list())

print("\nCleaned Ages (Invalids replaced with NaN):")
print(cleaned_ages.to_list())
```

```python
#Section E: Integrated Data Wrangling Exercise 23: Cleaning a Customer Dataset
import pandas as pd
import numpy as np
data = {
    "Student Name": ["Alice", "Bob", "Charlie", "Alice", "Diana", "Evan", "Bob"],
    "City": ["addis ababa", "ADDIS ABABA", " DIRE DAWA", "addis ababa", "Hawassa", "dire dawa", "ADDIS ABABA"],
    "Age": [20, np.nan, 19, 20, 22, np.nan, np.nan],
    "Score": [85.5, 92.0, np.nan, 85.5, 88.0, 45.0, 92.0]
}

df = pd.DataFrame(data)
print("--- Original Messy DataFrame ---")
print(df)
print("\n" + "="*50 + "\n")

df['City'] = df['City'].str.strip().str.title()

df = df.drop_duplicates()

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Score'] = df['Score'].fillna(df['Score'].median())

print("--- Cleaned DataFrame ---")
print(df)
```

```python
#Section E: Integrated Data Wrangling Exercise 23, Outlier Detection Using IQR
import pandas as pd
data = {"Sales": [120, 150, 170, 140, 160, 130, 150, 1100, 145, 135, 125, 155, 15, 165]}
df = pd.DataFrame(data)
Q1=df['Sales'].quantile(0.25)
Q3=df['Sales'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5 *IQR
upper_limit=Q3+ 1.5 *IQR
outliers=df[(df['Sales'] <lower_limit)| (df['Sales'] >upper_limit)]
print(outliers)
print(lower_limit)

print(df)
```
