```python
# import ing neccessary libraries
import requests
from datetime import datetime
from bs4 import BeautifulSoup
import pandas as pd
# Decalring nceccessary varibales
url="https://www.scrapethissite.com/pages/simple/"
#Avoiding some blocks
headers ={"User-agent":"Mozilla/5.0"}
#Integrating the url and headers to BeautifulSoup
response=requests.get(url, headers=headers)
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
# print(response)
# Checking the response
if response.status_code==200:
    print("Connection successful")
    soup=BeautifulSoup(response.text, 'html.parser')
    ## selecting country by className
    country_boxes = soup.find_all('div', class_='country')
    # print(country_boxes)
    all_countries=[]
    for country in country_boxes:
        ## Findiing each elemnt of country's description
        name_element=country.find('h3', class_='country-name')
        population_element=country.find('span', class_='country-population')
        capital_element=country.find('span', class_='country-capital')
        area_element=country.find('span', class_='country-area')
        # print(name_element)
        ## Makeing the elements clean, grap only the texts
        name=name_element.text.strip() if name_element else "N/A"
        capital=capital_element.text.strip() if capital_element else "N/A"
        population=int(population_element.text.strip()) if population_element else 0
        area=float(area_element.text.strip()) if area_element else 0.0
        country_info={
            "Country Name":name,
            "Country Capital":capital,
            "Country Population":population,
            "Country Area": area,
        }
        all_countries.append(country_info)
        # print(all_countries)
        # defining the dataframe to set the scapped data  to data in local storage
    df =pd.DataFrame(all_countries)
    df.to_csv(f'countries_{timestamp}.csv', index=False)
    df.to_json(f'countries_{timestamp}.json', orient='records', indent=2)
    df.to_excel(f'countries_{timestamp}.xlsx', index=False)

    print(f"Saved {len(df)} countries in 3 formats:")
    print(f"  ✅ countries_{timestamp}.csv")
    print(f"  ✅ countries_{timestamp}.json")
    print(f"  ✅ countries_{timestamp}.xlsx")
    print(f"Most populated country: {df.loc[df['Country Population'].idxmax(), 'Country Name']}")
    print(f"Largest country by area: {df.loc[df['Country Area'].idxmax(), 'Country Name']}")
else:
    print(f"Failed connection")
```
