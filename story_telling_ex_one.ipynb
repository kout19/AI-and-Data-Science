```python
#Exercise 1: Monthly Sales Trend Data
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

# Setting the plt size
plt.figure(figsize=(10,5))

# Setting the x and y axis values (ADDED label="Monthly Sales" here)
plt.plot(months, sales, marker="s", color="g", markersize=8, linestyle="--", linewidth=1, label="Monthly Sales")

plt.title("Monthly Sales", fontsize=13, fontweight="bold", fontfamily="arial")

# Setting the labels
plt.xlabel("months for a sales")
plt.ylabel("Sales for a month")
plt.annotate('Dragon appears!', xy=('The Midpoint Battle', 70),  =('Entering the Forest', 80),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(loc="upper left", fontsize=10)

# Setting the grid line on each row
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()


```

```python
#Exercise 2: Student Scores Data
import matplotlib.pyplot as plt

subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]
plt.figure(figsize=(10,5))
plt.plot(subjects, scores, marker="*", markersize="8", linestyle="--", linewidth="1", color="r", markerfacecolor="y", label="Scores")
plt.title("Exam Scores")
plt.legend(loc="upper left")
plt.show()
```

```python
#Exercise 4: Fruit Supply (Vertical Bar Chart)
import matplotlib.pyplot as plt

fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']
plt.figure(figsize=(10, 5))
plt.bar(fruits, counts,color=colors, edgecolor='black', linewidth=1, label="Bar chart")
plt.title("Fruit Supply")
plt.ylabel("Quantity")
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt

fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']

# 1. Setup the canvas (fig) and the plot box (ax)
fig, ax = plt.subplots(figsize=(8, 5))

# 2. Tell the specific "ax" box to draw the bars
ax.bar(fruits, counts, color=colors, edgecolor='black')

# 3. Use "ax.set_" to change titles and labels
ax.set_title("Fruit Counts Inventory", fontsize=14, fontweight='bold')
ax.set_xlabel("Fruit Types")
ax.set_ylabel("Quantity Available")
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
departments = ('CS', 'IS', 'SE', 'IT')
students = {
'Male': np.array([70, 65, 80, 50]),
'Female': np.array([60, 55, 70, 45])
}
fig, ax=plt.subplots(figsize=(10,5))
male_bars=ax.bar(departments, students['Male'], color="skyblue", label="Male", edgecolor="black", width=0.7)
female_bars=ax.bar(departments, students['Female'], bottom=students['Male'], color="salmon", label="Female",width=0.7, edgecolor="black")
ax.bar_label(female_bars, label_type="center", color="white", fontweight="bold", fontsize=11)
ax.bar_label(male_bars, label_type="center", color="white", fontweight="bold", fontsize=11)
ax.set_title("Studnets distribution by departmetn and gender", fontsize=13, fontweight="bold")

ax.set_xlabel("Departments", fontsize=12)
ax.set_ylabel("Number of students", fontsize=12)
ax.legend(loc="upper left")
ax.grid(axis='y', linestyle="--", alpha=0.5)
# ax.tight_layout()

plt.show()
```
