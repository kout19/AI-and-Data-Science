```python
import matplotlib.pyplot as plt
import seaborn as sns

#Firtst load datasets
tips=sns.load_dataset("tips")
print(tips)
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    hue="smoker",
    style="time",
    size="size",
    col="sex"
)
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    col="time"
)
plt.show()


    
    
    
```
